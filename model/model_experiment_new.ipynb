# Padding

```python
def encoded_result_pad(in_arr, length):
    """ encode result to 0 and 1
        eg [0,1,2,3] -> [0,1,1,1]
        and pad result to desire length
    """
    encoded = [0 if i == 0 else 1 for i in in_arr]
    pad_length = length - len(in_arr)
    return encoded + [0]*pad_length
```

```python
def pad_array(in_df, pad_row):
    """ pad 2d array to a specify dimension
        eg, original dimension is (5,5), pad_row = 7
            result dimension is (7,5), and extra rows are padded with 0s
    
    Parameters: in_df (DataFrame) - a dataframe
                pad_row (int) - pad row length
                
    Returns: (array) - padded array
    """
    my_arr = in_df.astype('float').values
    pad_length = pad_row - my_arr.shape[0]
    pad = np.zeros(pad_length * my_arr.shape[1]).reshape(pad_length, my_arr.shape[1])
    padded = np.row_stack([my_arr, pad])
    return padded
```

```python
for i,j in enumerate(subset):
    try:
        pad_array(j[main_col], 14)
    except:
        print(i)
```

```python
def drop_row_by_str(in_df, drop_if_str):
    """ drop rows if a str in that row
    """
    return in_df.drop([i for i,j in in_df.iterrows() if drop_if_str in j.values])
```

```python
subset = [drop_row_by_str(i, '-') for i in subset]
```

# 2d-convnet

```python
from keras import models
from keras import layers
from keras import optimizers
```

```python
model = models.Sequential()
model.add(layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu', input_shape=(14,11,1)))
model.add(layers.MaxPooling2D(pool_size=(2,2)))
model.add(layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2,2)))

model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(14, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
X_reshaped = np.array(X).reshape((1541,14,11,1))
```

```python
history = model.fit(X_reshaped, np.array(y), epochs=20)
```
